```python
import kitaevmodel
import numpy as np
```

```python
from kitaevmodel.shapes import (HexagonZigzag,
                                BandZigzag,
                                HexagonArmchair,
                                BandArmchair,
                                PeriodicSample,
                                HexagonHole,
                                HexagonZigzagLinSpec)
```

```python
from kitaevmodel.basesample import KitaevBase
```

```python
samples = [KitaevBase(8, 8, 0.027, 0.3, 0.1),
           HexagonZigzag(4, 0.027, 0.3, 0.1),
           BandZigzag(4, 6, 0.027, 0.3, 0.1),
           HexagonArmchair(4, 0.027, 0.3, 0.1),
           BandArmchair(4, 6, 0.027, 0.3, 0.1),
           PeriodicSample(8, 8, 0.027),
           HexagonHole(6, 2, 0.027, 0.3, 0.1),
           HexagonZigzagLinSpec(4, 0.027, 0.3, 0.1)]
```

```python
for sample in samples:
  sample.diagonalize()
  initial_state = np.zeros_like(sample.v[:, 0])
  initial_state[1] = 1
  sample.plot_state(initial_state, size=10)
  sample.animated_ev(initial_state, time=0.3, frames=5, size=10)
```
